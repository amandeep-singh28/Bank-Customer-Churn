```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_excel(r'Datasets\Bank_Churn_Messy.xlsx', sheet_name = 'Customer_Info')
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
df1['Surname'] = df1['Surname'].fillna(df1['Surname'].mode()[0])
df1['Age'] = df1['Age'].fillna(df1['Age'].mean())
```

```python
df1.isnull().sum()
```

```python
df1['Age'] = df1['Age'].astype('int64')
```

```python
df1.info()
```

```python
df1.head()
```

```python
df1['Geography'].unique()
```

```python
df1['Geography'] = df1['Geography'].replace('FRA', 'France')
df1['Geography'] = df1['Geography'].replace('French', 'France')
```

```python
df1['Geography'].unique()
```

```python
df1['EstimatedSalary'] = df1['EstimatedSalary'].str.replace('€','')
```

```python
df1.info()
```

```python
df1['EstimatedSalary'] = df1['EstimatedSalary'].astype('float64')
```

```python
df1.info()
```

```python
df1.head()
```

```python
df2 = pd.read_excel(r'Datasets\Bank_Churn_Messy.xlsx', sheet_name = 'Account_Info')
```

```python
df2.head()
```

```python
df2['Balance'] = df2['Balance'].str.replace('€','')
```

```python
df2['Balance'] = df2['Balance'].astype('float64')
```

```python
df2.head()
```

```python
df1.head()
```

```python
merged_df = pd.merge(df1, df2, on='CustomerId', suffixes=('_cust','_acc'))
```

```python
merged_df.head()
```

```python
merged_df = merged_df.drop(columns = ['Tenure_cust'])
```

```python
merged_df.rename(columns = {
    'Tenure_acc' : 'Tenure'
}, inplace = True)
```

```python
merged_df.head()
```

```python
merged_df.shape
```

```python
merged_df.nunique()
```

```python
merged_df.drop_duplicates(inplace = True)
```
